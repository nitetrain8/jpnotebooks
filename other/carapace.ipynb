```python
def ts2hms(t):
    t = t / 1000
    h = t // 3600
    m = (t-3600*h) // 60
    s = (t-3600*h-60*m)
    ms = (s - int(s))*1000
    return "%d:%.2d:%.2d.%.3d" %(h,m,s, ms)

def hms2(t):
    t = t / 1000
    h = t // 3600
    m = (t-3600*h) // 60
    s = (t-3600*h-60*m)
    ms = (s - int(s))*1000
    return "%.2d:%.2d.%.3d" %(m,s, ms)
```

```python
def events_near(events, ev, before, after):
    rv = []
    b = ev.timestamp - before
    a = ev.timestamp + after
    for e in events:
        if e.timestamp >= b:
            if e.timestamp <= a:
                rv.append(e)
            else:
                break
    return rv
evnr = events_near

def clusterify(events, before=100, after=100):
    l = set()
    for e in events:
        l.add(tuple(events_near(events, e, before, after)))
    l = [(c[0].timestamp, c) for c in l]
    
    # debug sanity check
    for c in l:
        saw = False
        for e in events:
            if e in c:
                if saw == True:
                    raise ValueError("event in multiple clusters")
                saw = True
    
    return sorted(l, key=lambda t: t[0])

```
