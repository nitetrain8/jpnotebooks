```python
NAME = 0
OPERATOR = 1
PUSH = 2
POP = 3
ATTR = 4
NUMBER = 5
EOF = 6
EQUAL = 7
NOTEQUAL = 8
LESSTHAN = 9
GREATERTHAN = 10
LESSEQUAL = 11
GREATEREQUAL = 12
DOT = 13
E_SYNTAX = 14
PLUS = 15
MINUS = 16
DIV = 17
MULT = 18
NUMBER = 19

NUM2STR = [
    "NAME",
    "OPERATOR", 
    "PUSH", 
    "POP", 
    "ATTR", 
    "NUMBER", 
    "EOF", 
    "==", 
    "!=", 
    "<", 
    ">", 
    "<=", 
    ">=", 
    ".", 
    "E_SYNTAX", 
    "+", 
    "-", 
    "/", 
    "*",
    "NUMBER"
]
    

class StringIter():
    def __init__(self, s):
        self.i = -1
        self.n = -1
        self.s = s
        self.l = len(s)
    def nextc(self):
        self.i += 1
        if self.i >= self.l:
            self.i = self.l
            return EOF
        return self.s[self.i]
    def ungetc(self):
        self.i -= 1
    def endtok(self):
        return self.s[self.n:self.i+1]
    def starttok(self):
        self.n = self.i
        

def getc(s):
    return s.nextc()

def ungetc(s):
    return s.ungetc()
    
def maybe_name(c):
    if c == EOF: return False
    o = ord(c)
    if 97 <= o <= 122: return True
    if 65 <= o <= 90 : return True
    if c == "_": return True
    return False

def isnumberstart(c):
    if c == EOF: return False
    return c in "0123456789-."

def isnumber(c):
    if c == EOF: return False
    return c in "0123456789."

def parseop(s,c):
    s.starttok()
    c2 = getc(s)
    if c == "=":
        if c2 != "=":
            ungetc(s)
        op = EQUAL
    elif c == "<":
        if c2 == ">":
            op = NOTEQUAL
        elif c2 == "=":
            op = LESSEQUAL
        else:
            ungetc(s)
            op = LESSTHAN
    elif c == ">":
        if c2 == "=":
            op = GREATEREQUAL
        else:
            ungetc(s)
            op = GREATERTHAN
    elif c == ".":
        ungetc(s)
        op = DOT
    elif c == "!":
        if c2 != "=":
            op = E_SYNTAX
    elif c == "+":
        ungetc(s)
        op = PLUS
    elif c == "-":
        ungetc(s)
        op = MINUS
    elif c == "/":
        ungetc(s)
        op = DIV
    elif c == "*":
        ungetc(s)
        op = MULT
    return op, s.endtok()
        

def is_name(c):
    if c == EOF: return False
    return maybe_name(c) or (48 <= ord(c) <= 57)

def tokenize3(st):
    s = StringIter(st)
    stack = []
    while True:
        c = getc(s)
        if c == EOF:
            break
        elif c == " ":
            continue
        elif maybe_name(c):               # NAME
            s.starttok()
            while is_name(c):
                c = getc(s)
            ungetc(s)
            tok = s.endtok()
            stack.append((NAME, tok))
            
        elif isnumberstart(c):
            s.starttok()
            while isnumber(c):
                c = getc(s)
            ungetc(s)
            tok = s.endtok()
            stack.append((NUMBER, tok))
            
        elif c in "({[":                  # PUSH
            stack.append((PUSH, c))
        elif c in ")}]":                  # POP
            stack.append((POP, c))
        elif c in "<>=+-/*.":
            v, tok = parseop(s, c)
            if v == E_SYNTAX:
                raise SyntaxError("Error at position %s"%s.i)
            stack.append((v, tok))
        else:
            raise SyntaxError(c)
    return stack
            
            
```
