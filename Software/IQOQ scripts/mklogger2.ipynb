```python
import os, clipboard
from collections import OrderedDict
pj = os.path.join
refd = "D:\\auto_hd_install\\default configs\\IM00226 Rev C\\Mag 03"
lsf = pj(refd, "Logging Off.log")
ls = set()
settings = OrderedDict()
with open(lsf, 'r') as f:
    f.readline()  # header
    txt = f.read().splitlines()
for line in txt:
    var, db, log, group = line.split("\t")
    ls.add(var)
    settings[var] = [db, log, group]
```

```python
def validate(v):
    if v not in ls:
        cs = "\n".join("%-40s: %.1f%%"%(a, b*100) for a,b in closest(v))
        raise ValueError("Invalid variable '%s'. Closest: \n\n%s"%(v, cs))

def closest(v):  # original
    l = ls
    vl = len(v)
    for i in range(len(v)):
        l2 = set()
        c = v[i]
        if v in ls:
            return [v]
        for var in l:
            try:
                c2 = var[i]
            except IndexError:
                continue
            if c == c2:
                l2.add(var)
        if not l2:
            break
        l = l2
    return sorted(l)

import difflib
def ratio(a,b):
    return difflib.SequenceMatcher(None, a,b).ratio()

def closest(v, n=5):
    scores = []
    for var in ls:
        s = ratio(v, var)
        scores.append((s, var))
    scores = sorted(scores, reverse=True)[:n]
    return [(b,a) for a,b in scores]
                
```

```python
import copy

class Logger():
    def __init__(self, name):
        self.name = name
        self.logging = []
        self.settings = copy.deepcopy(settings)
        
    def log(self, v, db):
        validate(v)
        self.logging.append((v, str(db)))
        
    def finish(self):
        for v, db in self.logging:
            s = self.settings[v]
            s[0] = db
            s[1] = "TRUE"
        out = ["Data Name\tThreshold change to record\tRecord\tGroup"]
        for v, (db, l, g) in self.settings.items():
            out.append("\t".join((v, db, l, g)))
        return "\n".join(out)
            
def logger(name):
    global __logger, log
    __logger = Logger(name)
    log = __logger.log

def finish(copy=False):
    s = __logger.finish()
    print(s)
    if copy:
        clipboard.copy(s)
    return s
```

```python
import contextlib
@contextlib.contextmanager
def mklogger(name, copy=False):
    logger(name)
    yield
    finish(copy)
    
```

```python
def finish(copy=False):
    s = __logger.finish()
    print(s)
    if copy:
        clipboard.copy(s)
    d = "D:\\IQOQ_3L_Mag\\Config"
    fp = pj(d, __logger.name)
    if not fp.endswith(".log"):
        fp += ".log"
    with open(fp, 'w') as f:
        f.write(s)
    return s
```
