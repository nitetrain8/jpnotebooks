```python
_bracket_find = re.compile(r"(.*?)\[(.*)\](.*)").findall
_ref_find = re.compile(req_item).findall

def extract_refs(lines):
    refs = []
    text = []
    if isinstance(lines, str):
        lines = lines.split()
    for line in lines:
        for before, bracket, after in _bracket_find(line):
            text.append(before)
            text.append(after)
            for typ, num in _ref_find(bracket):
                refs.append((typ, num))
    return refs, ''.join(text)
```

```python
req_item = "([FU]RS|SDS)([\d\.]+)"
_req_match = re.compile(req_item).match
_item_match = re.compile(r"^\>?[\*]*\s*[\+\*]+(?:%s)\:?[\+\*]*\:?\s*(.*?)\s*$" % req_item).match
_canceled_match = re.compile(r"^\>?[\*]*\s*-[\+\*]+(%s)\:?[\+\*]*\:?\s*(.*)\s*-$" % req_item).match
_header_match = re.compile(r"^\>?[\+\*]{2}([\d\.\w]+)[\+\*]{2}").match

def _extract_frs_lines(lines):
    start_matches = (
        (_item_match, 0),
        (_canceled_match, 1),
    )
    end_matches = (
        _item_match,
        _canceled_match,
    )
    i = 0
    while True:
        try:
            line = lines[i].strip()
        except IndexError:
            break
        i += 1
        for match, flags in start_matches:
            done = False
            if not line:
                m = None
            else:
                m = match(line)
            if m:
                req_typ, num, text1 = m.groups()
                req_lines = [text1]
                refs = []
                while True:
                    try:
                        line = lines[i].strip()
                    except IndexError:
                        done = True
                        break
                    if not line:
                        done = True
                        break
                    if any(m(line) for m in end_matches):
                        done = True
                        break
                    
                    req_lines.append(" " + line.strip())
                    i += 1
                refs, text = extract_refs(req_lines)
                yield req_typ, num, flags, refs, text
            if done:
                break          
```

```python
order = [
    'URS',
    'FRS',
    'SDS',
    'IQ',
    'OQ',
    'PQ'
]
    
class Requirement():
    def __init__(self, typ, num="", obs=False, refs=(), text="", _notag=False):
        self.type = typ
        self.num = num
        self.obs = obs
        self.refs = refs
        self.text = text
        if _notag:
            self.tag = ""
        else:
            self.tag = typ + num
        
        self.parents = set()
        self.children = set()
        
    def __hash__(self):
        return hash(self.tag)
    
    def childify(self):
        return self._childify([])
        
    def _childify(self, path):
        path.append(self)
        if not self.children:
            yield path
        else:
            for child in self.children:
                yield from child._childify(path.copy())
                
    def __repr__(self):
        return ("%s('%s%s')"%(self.__class__.__name__, self.type, self.num))
    
    def rowify(self):
        for row in self.childify():
            yield [r.tag for r in row]
                
    def memfree(self):
        for c in self.children:
            c.memfree()
        self.children.clear()
        self.parents.clear()

def setparent(child, parent):
    child.parents.add(parent)
    parent.children.add(child)
    
def empty_parent(tag):
    typ = empty_parents[tag]
    return data[typ][0]
        
# Build maps of tag -> Requirement and typ -> list of reqs for that type
# The "empty" parent items are used to output blank ("") cells for rows
# with no parents, e.g. an FRS item with no references to a URS

reqs = {}           # tag -> Requirement
data = {}           # typ -> List[Requirement]
empty_parents = {}  # typ -> empty parent Requirement
empty_children = {} # typ -> empty child Requirement
DUMMY_PARENT = parent = Requirement("dummy_parent")

for typ in order:
    r = Requirement(typ, _notag=True)
    setparent(r, parent)
    data[typ] = [r]
    empty_parents[typ] = parent
    parent = r
    
for i in range(len(order) - 1):
    empty_children[order[i]] = data[order[i + 1]][0]
    
del parent, typ, r

def empty_parent(typ):
    return empty_parents[typ]

def empty_child(typ):
    return empty_children[typ]

# 1st pass - build requirements data
for typ, num, obsolete, refs, text in _extract_frs_lines(raw.splitlines()):
    refs = set("".join(r) for r in refs)
    req = Requirement(typ, num, obsolete, refs, text)
    reqs[req.tag] = req
    data[req.type].append(req)
del typ, num, obsolete, refs, text, req
    
# second pass - with full flat list known, create references
# to parents and children by looking up parent references.
# For items with no parents, map to an empty parent
for req in reqs.values():
    if not req.refs:
        setparent(req, empty_parent(req.type))
        
    for ref in req.refs:
        parent = reqs.get(ref)
        if parent is None:
            raise ValueError("Can't find item: '%s'"%ref)
        setparent(req, parent)

_last_item = order[-1]
for req in reqs.values():
    if not req.children and req.type != _last_item:
        req.children.add(empty_child(req.type))
del req, parent, ref, _last_item
    
rows = []
for ur in data['URS']:
    rows.extend(ur.childify())

#sortrows2(rows)
partial_sort(rows, 0, key=lambda s: [int(n or 0) for n in s.num.split(".")])
partial_sort(rows, 1, key=lambda s: [int(n or 0) for n in s.num.split(".")])

rows2 = []
for row in rows:
    rows2.append([r.tag for r in row])

# first row *should* be the entirely blank row
# consisting of all empty requirement objects
assert rows2[0] == [""] * len(order), rows2[0]
del rows2[0] 

for row in rows2:
    print(row)
```
