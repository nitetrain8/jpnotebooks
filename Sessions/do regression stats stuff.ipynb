```python
ss = cr("R107")
m1c, m2c, bc, m3c = cr("D74:G74")
```

```python
import random
def pm10(c):
    return c.Offset(0,1).Value

def rv(v, r=0.1):
    if r == 0:return v
    i = 1+r
    j = 0.5/r
    return v*(i-random.random()/j)

m1v = pm10(m1c)
m2v = pm10(m2c)
bv = pm10(bc)
m3v = pm10(m3c)

def getrv(v, n=100, r=0.1):
    return [v] + [rv(v,r) for _ in range(n)]

def getrv(v, n=100, r=0.1):
    return np.linspace((1-r)*v, (1+r)*v, n)

def setv(c,v):
    c.Value=v
def getv(c):
    return c.Value
```

```python
def delmax(s, l=10):
    if len(s) < l:
        return
    m = None
    tm = None
    for t in s:
        v = t[0]
        if m is None or v > m:
            m = v
            tm = t
    s.remove(tm)
```

```python
def test(n1=100, n2=100, n3=100, n4=100, r=0.1):
    best = set()
    highest = None
    for m1 in getrv(m1v, n1, r):
        setv(m1c, m1)
        for m2 in getrv(m2v, n2, r):
            setv(m2c, m2)
            for b in getrv(bv, n3, r):
                setv(bc, b)
                for m3 in getrv(m3v, n4, r):
                    setv(m3c, m3)
                    ssv = getv(ss)
                    if highest is None or highest > ssv or len(best) < 10:
                        highest = ssv
                        delmax(best,10)
                        best.add((ssv, m1, m2, b, m3))
    be = sorted(best)
    _, m1, m2, b, m3 = be[0]
    setv(m1c, m1)
    setv(m2c, m2)
    setv(bc, b)
    setv(m3c, m3)
    return be
```

```python
import time
def ttest(f,*args):
    t = time.time
    t1 = t()
    rv=f(*args)
    t2 = t()
    el = t2-t1
    print("%s: %.2f seconds elapsed"% (f.__name__, el))
    return el, rv
```

```python
with screen_lock(xl):
    res = ttest(test, 1000, 0,0,0, 0.1)
for r in res:
    print(r)
```

```python
with screen_lock(xl):
    res = ttest(test, 10, 10,10,10, 0.1)
for r in res:
    print(r)
```

```python
refdata = cr("Q71:V75").Value
rd = []
for i, row in enumerate(refdata):
    o2 = 20*(i+1)
    for j,c in enumerate(row):
        mg = j / 10
        rd.append((o2, mg, c))
rdkla = [r[2] for r in rd]
ref = np.array(rdkla)
```

```python
import numpy as np
m1v, m2v, bv, m3v = cr("D74:G74").Value[0]
mgs = []
o2s = []
for i in range(5):
    o2 = 20*(i+1)
    for j in range(6):
        mg = j / 10
        o2s.append(o2)
        mgs.append(mg)
mgs = np.array(mgs)
o2s = np.array(o2s)
mxo = mgs*o2s
```

```python
def getk(mg, o2, mxo=None, m1=m1v, m2=m2v, b=bv, m3=m3v):
    if mxo is None: mxo=mg*o2
    return m1*mxo + b + m2*mg + m3*o2

def sumss(a,b):
    return np.sum((a-b)**2)

def getss(m1=m1v, m2=m2v, b=bv, m3=m3v):
    k = getk(mgs, o2s, mxo, m1, m2, b, m3)
    return sumss(ref, k)*100000

def lin_iter(n1=10, n2=10, n3=10, n4=10, r=0.1):
    for a in getrv(m1v, n1, r):
        for b in getrv(m2v, n2, r):
            for c in getrv(bv, n3, r):
                for d in getrv(m3v, n4, r):
                    yield a,b,c,d
    
```

```python
def test2(n=10):
    best = set()
    highest = None
    for m1, m2, b, m3 in lin_iter(n, n, n, n):
        ssv = getss(m1, m2, b, m3)
        if highest is None or highest > ssv or len(best) < 10:
            highest = ssv
            delmax(best,10)
            best.add((ssv, m1, m2, b, m3))
    return sorted(best)
```

```python
dat = []
rs = []
for n in (1, 2, 3, 5, 7, 10, 15, 20, 40):
    t, res=ttest(lambda:test2(n))
    dat.append((n,t))
    rs.extend(res)
sorted(res)
```
