```python
import os, shutil
```

```python

def copy_for_v3():
    # copy script for v3.0
    fp = 'E:\\Install\\PBS Biotech Desktop\\1 Install RIO'
    out = 'D:\\auto_hd_install\\default configs\\IM00226 Rev C'
    os.makedirs(out, exist_ok=True)
    files = []
    for d in os.listdir(fp):
        ff = os.path.join(fp, d)
        if os.path.isdir(ff) and d.split()[0] in ("Air", "Mag"):
            tdir = os.path.join(out, d)
            os.makedirs(tdir, exist_ok=True)
            for d in os.listdir(os.path.join(ff, "Config")):
                if d == 'yRaw':
                    tdir = os.path.join(tdir, "yRaw")
                    os.makedirs(tdir, exist_ok=True)
                for cf in os.listdir(os.path.join(ff, "Config", d)):
                    ffp = os.path.join(ff, "Config", d, cf)
                    #print("Copying %r to %r"%(cf, tdir))
                    files.append((ffp, os.path.join(tdir, cf)))
    for i, (a,b) in enumerate(files,1):
        c1, _, _, c3 = a.split("\\")[-4:]
        c = "\\".join((c1, c3)) # os.path.basename(a)
        d = "\\".join(b.split("\\")[-2:])
        print("File %d/%d: %s to %s"%(i, len(files), c, d))
        shutil.copy2(a,b)
```

```python
copy_for_v3()
```
