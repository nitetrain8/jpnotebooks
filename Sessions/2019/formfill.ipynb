```python
#!/usr/bin/env python3

import sys
import os

from pdfrw import PdfReader, PdfWriter
from pdfrw.findobjs import find_objects
from pdfrw.objects import PdfName

os.chdir("C:\\Users\\Nathan\\Documents\\PBS\\WIP procedures-reports\\Preventative Maintenance")

for annot in find_objects(
  PdfReader("test.pdf").pages, 
  valid_types=(PdfName.XObject,PdfName.Annot,), 
  valid_subtypes=(PdfName.Widget,)
):
  if (annot.FT == '/Btn'):
    AP = annot.AP

outfn = "test23.pdf"
inpfn = "test.pdf"
pdf = PdfReader(inpfn)
pages = pdf.pages
for annot in find_objects(pages, valid_types=(PdfName.XObject, None), valid_subtypes=(PdfName.Widget,)):
    if (annot.FT == '/Btn'):
        annot['/H'] = '/I'
        annot.AP['/H']= '/I'
        

writer = PdfWriter(outfn, trailer=pdf)
writer.write()
```
