```python
import clipboard
def fix(typ="uint"):
    out = []
    for line in clipboard.paste().splitlines():
        line = "public static readonly %s %s;" % (typ, line.lstrip().replace(",", ""))
        out.append(line)
    clipboard.copy("\n        ".join(out))
```
